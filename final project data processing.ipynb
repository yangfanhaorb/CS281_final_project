```python
# define some constants

reign_name_order = {u'崇德':0,u'顺治':1,u'康熙':2,u'雍正':3,u'乾隆':4,u'嘉庆':5,u'道光':6,u'咸丰':7,u'同治':8,u'光绪':9,u'宣统':10}
reign_name_start_year = {u'崇德':1636,u'顺治':1644,u'康熙':1662,
                         u'雍正':1723,u'乾隆':1736,u'嘉庆':1796,
                         u'道光':1821,u'咸丰':1851,u'同治':1862,
                         u'光绪':1875,u'宣统':1909}
```

```python
# useful function for year parsing

def extract_year(text):
    counter = 0
    while text[counter] != u'年':
        counter += 1
    return text[:counter]

def parse_year(chn_year):
    num_dict = {u'一':1,u'元':1,u'二':2,u'三':3,u'四':4,u'五':5,u'六':6,u'七':7,u'八':8,u'九':9,u'十':10,u'伍':5}
    year_elements = [num_dict[item] for item in chn_year]
    year = 0
    for num in year_elements:
        if num != 10:
            year += num
        else:
            if year == 0:
                year += 10
            else:
                year *= 10
    return year
```

```python
# define document class

class Document:
    def __init__(self, text, reign_name, reign_year):
        self.text = text
        self.reign_name = reign_name
        self.reign_year = reign_year
        self.year = reign_name_start_year[reign_name] + reign_year - 1
    def __str__(self):
        return self.reign_name.encode('utf8') \
            + str(self.reign_year) \
            +'年 (western calendar:'+str(self.year)+')\n' + self.text.encode('utf8')
    def get_text_length(self):
        return len(self.text)
```

```python
# pick out the documents and associate year markers
counter = 0
curr_reign_name = u'崇德'
curr_reign_year = 8

documents = []
for i in range(len(full_text)):
    try:
        if full_text[i][0] == u'○' or full_text[i][0] == u'□':
            documents.append(Document(full_text[i][1:], curr_reign_name, curr_reign_year))
            counter += 1
        else:
            if full_text[i][:2] in reign_name_order and (full_text[i+1][0] == u'○' or full_text[i][0] == u'□'):
                # change current year marker
                new_reign_name = full_text[i][:2]
                new_reign_year = parse_year(extract_year(full_text[i][2:]))
                if (new_reign_name == curr_reign_name) and (new_reign_year ==  curr_reign_year):
                    pass
                elif (new_reign_name == curr_reign_name) and (new_reign_year == curr_reign_year + 1):
                    curr_reign_year = new_reign_year
                elif (new_reign_name != curr_reign_name) \
                        and (reign_name_order[curr_reign_name] == reign_name_order[new_reign_name] - 1) \
                        and (new_reign_year == 1):
                    curr_reign_year = new_reign_year
                    curr_reign_name = new_reign_name
                # to account for some irregularities
                elif (new_reign_name == u'嘉庆') and (curr_reign_name == u'嘉庆') \
                        and (curr_reign_year == 4) and (new_reign_year == 1):
                    curr_reign_year = new_reign_year
                else:
                    print 'Inconsistencey in parsing year markers on line', i
                    print new_reign_name,new_reign_year
                    print curr_reign_name,curr_reign_year
    except:
        print 'Error on line',i
print counter
```

```python
# average document size
total_length = 0
for document in documents:
    total_length += document.get_text_length()
print 'Average document size is:',total_length/len(documents)
```

```python
# document size above a certain threshold 
document_longer_than_500 = 0
document_longer_than_200 = 0
document_longer_than_100 = 0
document_longer_than_50 = 0
document_longer_than_10 = 0
document_size_counts = [len(documents),0,0,0,0,0]
document_sizes = ['0','10','50','100','200','500']
for document in documents:
    size = document.get_text_length()
    if size >= 500:
        document_longer_than_500 += 1
        document_size_counts[5] += 1
    if size >= 200:
        document_longer_than_200 += 1
        document_size_counts[4] += 1
    if size >= 100:
        document_longer_than_100 += 1
        document_size_counts[3] += 1
    if size >= 50:
        document_longer_than_50 += 1
        document_size_counts[2] += 1
    if size >= 10:
        document_longer_than_10 += 1
        document_size_counts[1] += 1
print 'Number of documents in total:',len(documents)
print 'Number of documents longer than 500:', document_longer_than_500
print 'Number of documents longer than 200:', document_longer_than_200
print 'Number of documents longer than 100:', document_longer_than_100
print 'Number of documents longer than 50:', document_longer_than_50
print 'Number of documents longer than 10:', document_longer_than_10
```
