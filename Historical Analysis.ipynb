```python
# load the topic composition

topic_composition = []
with open('mingshilu_2_gram_composition.txt', 'r') as f:
    for line in f:
        to_add = line.split('\t')
        topic_composition.append(to_add)
```

```python
# extract the proportion of topics on Japanese pirates for each year
from collections import defaultdict

japanese_pirates_per_year = defaultdict(list)
look_up_document = dict()
for document in topic_composition:
    year = int(document[1][-8:-4])
    japanese_pirates = float(document[8])
    japanese_pirates_per_year[year].append(japanese_pirates)
    look_up_document[japanese_pirates] = document[0]
```

```python
# average the proportion for each year
japanese_pirates_per_year_rate = defaultdict(float)

for year in japanese_pirates_per_year.keys():
    japanese_pirates_per_year_rate[year] = sum(japanese_pirates_per_year[year])/len(japanese_pirates_per_year[year])
```

```python
# plot graph for Japanese pirates

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
ind = np.arange(1350,1650)
x = []
for i in ind:
    x.append(japanese_pirates_per_year_rate[i])
plt.plot(ind,x)
plt.xlabel('Year')
plt.ylabel('Topic Proportion')
plt.rcParams['figure.figsize']=(20,10) 
plt.title('Proportion of Topic Japanese Pirates by Year for Mingshilu')
plt.savefig('japanese_pirates_mingshilu_by_year.png')
```
